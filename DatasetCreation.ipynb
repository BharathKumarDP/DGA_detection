```python
import pandas as pd
import os 
import re
from pathlib import Path
```

```python
rootdir = "../DGA_Main/"
regex = re.compile('10000.txt')

DGA_domains = []
DGA_domain_class = []
is_DGA = []
for root, dirs, files in os.walk(rootdir):
    for file in files:
        if file == "10000.txt":
            fullpath = os.path.join(root, file)
            with open(fullpath) as dga_file:
                lines = dga_file.readlines()
                lines = [line.rstrip() for line in lines]
            class_name = os.path.split(os.path.split(root)[0])[1]
            print(class_name)
            DGA_domain_class.extend([class_name] * len(lines))
            DGA_domains.extend(lines)
            is_DGA.extend([0] * len(lines))
```

```python
print(len(DGA_domains))
print(len(DGA_domain_class))
print(len(is_DGA))
```

```python
norm_domains = pd.read_csv('../DGA datasets/majestic_million_csv.csv')
```

```python
norm_domains.head()
```

```python
norm_domain_list = list(norm_domains['Domain'].values[:len(DGA_domains)])
norm_domain_class = ["None"]*len(DGA_domains)
is_DGA2 = [1]*len(DGA_domains)
```

```python
print(len(norm_domain_list))
print(len(norm_domain_class))
print(len(is_DGA2))
```

```python
dataset = pd.DataFrame(list(zip(DGA_domains+norm_domain_list, is_DGA + is_DGA2, DGA_domain_class+norm_domain_class)), columns = ['Domain', 'Label', 'DGA Family'])
```

```python
dataset.to_csv('dataset.csv', index=False)
```
