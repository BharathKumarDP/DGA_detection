```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN 
```

```python
data=pd.read_csv('dataset_7_features.csv')
data=np.array(data)
#features
X=data[:,3:]
#binary label
y=data[:,1]
y=y.astype('int')
```

```python
y
```

```python
class_labels=data[:,2]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
c_label = le.fit_transform(class_labels)
c_label
```

```python
cnt=0
prev=c_label[0]
for i in range(len(c_label)):
    print(c_label[i])
    if c_label[i]!=prev:
        cnt+=1
        prev=c_label[i]
    if cnt==5:
        print(i)
        break
```

```python
#taking first 5 classes,(4,6,8,1,2)till index 50,000
X_5=X[:50000,]
c_label_5=c_label[:50000]
len(X_5),len(c_label_5)
```

```python
#train,test,split
#test_size
#valid_size
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, Y_train, Y_val = train_test_split(X_train,Y_train, test_size=0.1)
```

```python
#Dimensionality reduction
#1.PCA
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(X_5)
```

```python
#reducing to 3 features which explains more than 90% variance
pca=PCA(n_components=2)
reduced_X=pca.fit_transform(X_5,y)
reduced_X.shape
```

```python
#reducing to 2 features which explains more than 90% variance
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda_2= LinearDiscriminantAnalysis(n_components=2)
reduced_X=lda_2.fit_transform(X[:50000],c_label[:50000])
reduced_X.shape
```

```python
db = DBSCAN(eps=0.2, min_samples=20).fit(reduced_X)
labels = db.labels_
fig = plt.figure(figsize=(10, 10))
sns.scatterplot(reduced_X[:,0], reduced_X[:,1], hue=["cluster-{}".format(x) for x in labels])
```

```python
db = DBSCAN(eps=0.5, min_samples=4).fit(reduced_X)
labels = db.labels_
fig = plt.figure(figsize=(10, 10))
sns.scatterplot(reduced_X[:,0], reduced_X[:,1], hue=["cluster-{}".format(x) for x in labels])
```

```python
from sklearn.neighbors import NearestNeighbors
nearest_neighbors = NearestNeighbors(n_neighbors=15)
neighbors = nearest_neighbors.fit(X)
distances, indices = neighbors.kneighbors(X)
distances = np.sort(distances[:,10], axis=0)
fig = plt.figure(figsize=(5, 5))
plt.plot(distances)
plt.xlabel("Points")
plt.ylabel("Distance")
plt.savefig("Distance_curve.png", dpi=300)
```

```python
fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.5, wspace=.2)
i = 1
for x in range(20, 0, -1):
    eps = 1/(21-x)
    db = DBSCAN(eps=eps, min_samples=20).fit(reduced_X)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_
    
    print(eps)
    ax = fig.add_subplot(2, 5, i)
    ax.text(1, 4, "eps = {}".format(round(eps, 1)), fontsize=25, ha="center")
    sns.scatterplot(X[:,0], X[:,1], hue=["cluster-{}".format(x) for x in labels])
    
    i += 1
```
