```python
def remove_tlds(domain):
    words=domain.split(".")
    return words[0]
```

```python
def length(domain):
    return len(domain)
```

```python
def get_freq(domains):
    char_freq={}
    tot=0
    for i in range(len(domains)):
        count=Counter(domains[i])
        for key in count:
            if key not in char_freq:
                char_freq[key]=0
            char_freq[key]+=count[key]
            tot+=count[key]

    len(char_freq)
    
    for key in char_freq:
        char_freq[key]=(char_freq[key]/tot)
        
    return char_freq
```

```python
def relative_entropy(data, base):
    entropy = 0.0
    length = len(data) * 1.0
    
    if length > 0:
        cnt= Counter(data)
        
    for char, count in cnt.items():
        observed = count / length
        expected = base[char]
        entropy += observed * math.log((observed / expected),2)
        
    return entropy
```

```python
def count_num(domain):
    cnt=0
    for let in domain:
        if let.isnumeric():
            cnt=cnt+1
    
    return cnt/len(domain)
```

```python
def remove_substr(int_words):
    copy_words=int_words.copy()
    for sub in copy_words:
        subs={sub[i:j+i] for j in range(2,len(sub)) for i in range(len(sub)-j+1)}
        for i in subs:
            if i in int_words:
                int_words.remove(i)
    
    return int_words
```

```python
def jaccard_index(m_domain,domains):
    jaccard=0
    for dom in domains:
        abs_ins=np.intersect1d(m_domain,dom).shape[0]
        abs_uni=np.union1d(m_domain,dom).shape[0]
        jaccard+=(abs_ins/abs_uni)*1.0

    return (jaccard*1.0)/len(domains)
```

```python
def max_consecutive_consonants(domain):
    consonant_list = re.findall(r'[bcdfghjklmnpqrstvwxyz]+', domain , re.IGNORECASE)
    if len(consonant_list) > 0:
        return len(max(consonant_list, key=len))
    else:
        return 0
```

```python
def max_consecutive_vowels(domain):
    vowel_list = re.findall(r'[aeiou]+', domain , re.IGNORECASE)
    if len(vowel_list) > 0:
        return len(max(vowel_list, key=len))
    else:
        return 0
```

```python
def vowel_count(domain):
    return len(re.findall(r'[aeiou]', domain , re.IGNORECASE))
```

```python
def vowel_rate(domain):
    return len(re.findall(r'[aeiou]', domain , re.IGNORECASE))/len(domain)
```
