```python
import numpy as np
import pandas as pd
import math
```

```python
domains=[]
data=pd.read_csv('DGA datasets/DGA datasets/majestic_million_csv.csv')
m_data=pd.read_csv('murofet_v1/murofet.csv')
```

```python
def remove_tlds(domains):
    for i in range(len(domains)):
        words=domains[i].split(".")
        domains[i]=words[0]
        if len(words)>2:
            print(words)
    return domains
```

```python
#fetching domain names
data=np.array(data)
data
domains=data[:,2]
domains=remove_tlds(domains)
domains.shape

m_data=np.array(m_data)
m_domains=m_data[:,0]
m_domains=remove_tlds(m_domains)
m_domains
```

```python
from collections import Counter

def get_freq(domains):
    char_freq={}
    tot=0
    for i in range(len(domains)):
        count=Counter(domains[i])
        for key in count:
            if key not in char_freq:
                char_freq[key]=0
            char_freq[key]+=count[key]
            tot+=count[key]

    len(char_freq)
    
    for key in char_freq:
        char_freq[key]=(char_freq[key]/tot)
        
    return char_freq
```

```python
def relative_entropy(base,data):
    entropy = 0.0
    length = len(data) * 1.0
    
    if length > 0:
        cnt= Counter(data)
        
    for char, count in cnt.items():
        observed = count / length
        expected = base[char]
        entropy += observed * math.log((observed / expected),2)
        
    return entropy
```

```python
#computes bw a m_domain and all r_domainsTODO:with predecessor
def jaccard_index(m_domain,domains):
    jaccard=0
    for dom in domains:
        abs_ins=np.intersect1d(m_domain,dom).shape[0]
        abs_uni=np.union1d(m_domain,dom).shape[0]
        jaccard+=(abs_ins/abs_uni)*1.0
    
    print(jaccard)
    return (jaccard*1.0)/len(domains)
```

```python
#computes bw a m_domain and all r_domainsTODO:with predecessor
def edit_distance(m_domain,domains):
    score=0
    for dom in domains:
        score+=enchant.utils.levenshtein(m_domain,dom)
        
    return score/len(domains)
```

```python
def count_num(domain):
    cnt=0
    for let in domain:
        if let.isnumeric():
            cnt=cnt+1
    
    return cnt/len(domain)


def length(domain):
    return len(domain)
```

```python
#preparing list of english words
file1=open('english_words_479k.txt')
eng=file1.readlines()
eng_vocab=[]
for words in eng:
    eng_vocab.append(words.strip('\n'))
eng_vocab
```

```python
def meaning_ratio(dom,eng_vocab):
    m_ratios=[]
    sub_str={dom[i:j+i] for j in range(2,len(dom)) for i in range(len(dom)-j+1)}
    sub_str=list(sub_str)
    int_words=list(np.intersect1d(sub_str,eng_vocab))
    int_words=remove_substr(int_words)
    tot=0
    for words in int_words:
        tot+=len(words)
    print(tot/len(dom))
    m_ratios.append(tot/len(dom))
    
    return tot/len(dom)
```

```python
def remove_substr(int_words):
    copy_words=int_words.copy()
    for sub in copy_words:
        subs={sub[i:j+i] for j in range(2,len(sub)) for i in range(len(sub)-j+1)}
        for i in subs:
            if i in int_words:
                int_words.remove(i)
    
    return int_words
```

```python
fields=['domain','r_entropy','numerical_percentage','length','meaning_ratio','jaccard_index']
records=[]
for dom in m_domains:
    record=[]
    record.append(dom)
    char_freq=get_freq(m_domains)
    record.append(relative_entropy(char_freq,dom))
    record.append(count_num(dom))
    record.append(length(dom))
    record.append(meaning_ratio(dom,eng_vocab))
    record.append(jaccard_index(dom,domains))

import csv
filename = "prep.csv"
     
with open(filename, 'w') as csvfile:  
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerow(fields) 
    csvwriter.writerows(records)
    
```
